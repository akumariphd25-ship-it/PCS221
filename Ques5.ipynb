```python
import numpy as np
```

```python
def create_even_corner_matrix(n):
    
    matrix = np.random.randint(0, 50, size=(n, n))
    
    # Force all elements to be even
    matrix = matrix * 2
    
    return matrix
```

```python
n = 5
matrix = create_even_corner_matrix(n)

print("Generated Matrix:\n")
print(matrix)
```
