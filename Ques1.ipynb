```python
import random
import csv
```

```python
filename = "random_database.csv"

rows = 20000
cols = 10000
```

```python
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)

    for r in range(rows):
        row_data = []

        for c in range(cols):
            # Different random generators
            if c % 5 == 0:
                value = random.randint(0, 1000)
            elif c % 5 == 1:
                value = random.uniform(0, 1000)
            elif c % 5 == 2:
                value = random.gauss(50, 15)
            elif c % 5 == 3:
                value = random.random()
            else:
                value = random.choice([0, 1])

            row_data.append(value)

        writer.writerow(row_data)

        # Progress update
        if r % 100 == 0:
            print(f"{r} rows written...")

print("Data file generation complete.")
```
