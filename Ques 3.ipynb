```python
import numpy as np
from collections import defaultdict
```

```python
# Generate 100 random integers between 0 and 99
data = np.random.randint(0, 100, 100)

data
```

```python
groups = defaultdict(list)

for value in data:
    group_key = value 
    groups[group_key].append(value)

groups
```

```python
# Find largest group size
max_length = max(len(group) for group in groups.values())


matrix = np.full((len(groups), max_length), np.nan)

# Fill matrix with grouped values
for row_index, key in enumerate(sorted(groups.keys())):
    matrix[row_index, :len(groups[key])] = groups[key]

matrix
```

```python
flat_clean = matrix[~np.isnan(matrix)]

flat_clean
```
